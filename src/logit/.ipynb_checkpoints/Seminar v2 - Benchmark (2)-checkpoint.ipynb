```
import pandas as pd
df_images = pd.read_csv('images.csv', sep=',',header=None)
df_labels = pd.read_csv('labels.csv', sep=',',header=None)
```

```
# randomly split data into 80% train and 20% test set
trainX, testX, trainY, testY = \
  model_selection.train_test_split(df_images, df_labels, 
  train_size=0.8, test_size=0.2, random_state=4487)
```

## Multinomial Regression

```
# learn logistic regression classifier
mlogreg = linear_model.LogisticRegressionCV(Cs=logspace(-4,4,20), cv=5, n_jobs=-1,
            multi_class='multinomial', solver='lbfgs', verbose=2,)  
            # use multi-class and corresponding solver
mlogreg.fit(trainX, trainY)

# predict from the model
predY = mlogreg.predict(testX)

# calculate accuracy
acc = metrics.accuracy_score(testY, predY)
print("test accuracy=", acc)
```

## Multiclass SVM

```
msvm = multiclass.OneVsRestClassifier(svm.SVC(kernel='linear'))
# msvm = multiclass.OneVsRestClassifier(svm.SVC(kernel='rbf'))

# setup the parameters and run CV
paramgrid = {'estimator__C': logspace(-3,3,13)}
msvmcv = model_selection.GridSearchCV(msvm, paramgrid, cv=5, n_jobs=-1, verbose=True)
msvmcv.fit(trainX, trainY)
predY = msvmcv.predict(testX)

# calculate accuracy
acc = metrics.accuracy_score(testY, predY)
print("test accuracy=", acc)
```
